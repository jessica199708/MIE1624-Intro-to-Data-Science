```python
import pandas as pd
import re
import unicodedata
```

## 1 Data Cleaning

```python
sentiment_analysis_df = pd.read_csv('sentiment_analysis.csv')
sentiment_analysis_df
```

```python
US_Elections_df = pd.read_csv('US_Elections_2020.csv')
US_Elections_df 
```

### 1.1 Remove all html tags and attributes (i.e., /<[^>]+>/) 

```python
sentiment_text = sentiment_analysis_df['text']
US_Elections_text = US_Elections_df ['text']
```

```python
# Using a regex, compile the regex pattern and match them into original text
def remove_html(text):
    cleanr = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')
    cleantext = re.sub(cleanr, '', text)
    return cleantext
```

```python
sentiment_analysis_df['text'] = sentiment_text.apply(remove_html)
sentiment_analysis_df
```

```python
US_Elections_df['text'] = US_Elections_text.apply(remove_html)
US_Elections_df
```

### 1.2 Html character codes (i.e., &...;) are replaced with an ASCII equivalent

```python
# Normal form form for the Unicode string
# NFKD will apply the compatibility decomposition, i.e. replace all compatibility characters with their equivalents
# Encode create creates a byte string, which cannot be mixed with a regular string
# Decode can convert back to string again
def ASCII_equivalent(text):
    ASCII_text = unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8')
    return ASCII_text
```

```python
sentiment_analysis_df['text'] = sentiment_text.apply(ASCII_equivalent)
sentiment_analysis_df
```

```python
US_Elections_df['text'] = US_Elections_text.apply(ASCII_equivalent)
US_Elections_df
```

### 1.3 Remove all URLs

```python
def remove_URL(text):
    cleanr = re.compile(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*')
    cleantext = re.sub(cleanr, '', text)
    return cleantext
```

```python
sentiment_analysis_df['text'] = sentiment_text.apply(remove_URL)
sentiment_analysis_df
```

```python
US_Elections_df['text'] = US_Elections_text.apply(remove_URL)
US_Elections_df
```

### 1.4 All characters in the text are in lowercase.

```python
sentiment_analysis_df['text'] = sentiment_text.str.lower()
sentiment_analysis_df
```

```python
US_Elections_df['text'] = US_Elections_text.str.lower()
US_Elections_df
```
